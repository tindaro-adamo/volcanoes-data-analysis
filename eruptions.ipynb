```python
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
df_eruptions =  pd.read_csv('csv/eruptions_2020.csv')
df_eruptions
```

```python
df_eruptions.isnull().sum()
```

```python
df_volcanoes = pd.read_csv('volcanoes_cleaned.csv')
df_volcanoes.columns
```

```python
# Add country information from volcanoes dataframe to eruptions df
df_eruptions = pd.merge(df_eruptions,df_volcanoes[['Volcano Number','Country']], how='left')
```

```python
# Let's check if there are eruptions with Volcano Number not present in df_volcanoes
df_eruptions[df_eruptions['Country'].isnull()]['Volcano Name'].value_counts()
```

```python
# We can Discard eruptions with Unknown Source, the others are very few
df_eruptions = df_eruptions[df_eruptions['Volcano Name']!='Unknown Source']
df_eruptions[df_eruptions['Country'].isnull()]['Volcano Name'].value_counts()
```
